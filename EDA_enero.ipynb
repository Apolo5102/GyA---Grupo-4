# Tabla dinamica Ledger

```python
Ledger_copia = Ledger.copy()
```

```python
Ledger.columns
```

```python
Ledger_final = Ledger[['Region',' Tax Level','County', 'City','Tax Collected']]
Ledger_final.head()
```

```python
#Tabla dinamica de Ledger donde las filas se agrupan por Region, County y City, las columnas se definen por Tax Level y los valores son la sumatoria de Tax Collected
tabla_comparacion = Ledger_final.pivot_table(index=['Region','County','City'], columns=' Tax Level', values='Tax Collected', aggfunc='sum')
tabla_comparacion
```

```python
tabla_comparacion.to_parquet('tabla_comparacion.parquet', index=False)
```
