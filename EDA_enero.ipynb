**Se identifica como FSBP tiene la mayor cantidad de formatos incorrectos, teniendo como texto varias fechas y valores de taxes necesarios para el desarrollo del proyecto**

```python
#Correcion de fechas Duro_Last

Duro_last['PERIOD USED'] = pd.to_datetime(Duro_last['PERIOD USED'],errors= 'coerce')
```

```python
Duro_last.info()
```

```python
#Correccion de fecha period used para FSBP y valores numericos para los niveles de pago de impuestos

FSBP['Period Used'] = pd.to_datetime(FSBP['Period Used'],errors= 'coerce')
FSBP['Expiration Date'] = pd.to_datetime(FSBP['Expiration Date'],errors= 'coerce')
FSBP['State Tax'] = pd.to_numeric(FSBP['State Tax'],errors= 'coerce')
FSBP['County Tax'] = pd.to_numeric(FSBP['County Tax'],errors= 'coerce')
FSBP['City Tax'] = pd.to_numeric(FSBP['City Tax'],errors= 'coerce')
```

```python
FSBP.info()
```

```python
Gaco['Period Used'] = pd.to_datetime(Gaco['Period Used'],errors= 'coerce')
```

```python
Gaco.info()
```

```python
Duro_last.to_csv('Duro_last.csv', index= False, date_format= "%Y-%m-%d")
FSBP.to_csv('FSBP.csv', index= False, date_format= "%Y-%m-%d")
SES.to_csv('SES.csv', index= False, date_format= "%Y-%m-%d")
Gaco.to_csv('Gaco.csv', index= False, date_format= "%Y-%m-%d")
Negativos.to_csv('Negativos.csv', index= False, date_format= "%Y-%m-%d")
```
