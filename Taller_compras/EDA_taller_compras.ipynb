# Transformacion (Correccion de formatos, tratamiento de nulos, imputacion, replace, etc)

```python
Clientes.dropna(subset=['Nombre'], inplace=True)
```

```python
copia = Clientes.copy()
```

```python
Clientes['Fecha Compra']= pd.to_datetime(Clientes['Fecha Compra'],format= 'mixed', errors='ignore')
```

```python
Clientes['Fecha Compra'].value_counts()
```

```python
Clientes.info()
```

```python
#Reemplazar caracter % en las columnas para cada Nombre y Apellido por n
Clientes['Nombre'] = Clientes['Nombre'].str.replace('%', 'n')
Clientes['Apellido'] = Clientes['Apellido'].str.replace('%', 'n')
```

```python
#Funcion para normalizar columnas de texto con strip() y title()
def normalizar_texto(df, columna):
    df[columna] = df[columna].str.strip().str.title()
```

```python
normalizar_texto(Clientes, 'Nombre')
normalizar_texto(Clientes, 'Apellido')
```

```python
#Nombres y apellidos estandarizados
Clientes.head()
```

```python
Clientes['Cantidad'] = Clientes['Cantidad'].fillna(Clientes['Valor Total']/Clientes['Valor Unitario'])
```

```python
Clientes.info()
```

```python
electrodometicos = Clientes.groupby('Electrodoméstico')['Valor Unitario'].mean().reset_index()
electrodometicos
```

```python
valores_unitarios = {
    600000.0: 'Aspiradora',
    300000.0: 'Cafetera',
    500000.0: 'Horno eléctrico',
    1500000.0: 'Lavadora',
    250000.0: 'Licuadora',
    400000.0: 'Microondas',
    1800000.0: 'Nevera',
    150000.0: 'Plancha',
    1200000.0: 'Televisor',
    200000.0: 'Ventilador'
}
```

```python
Clientes['Electrodoméstico'] = Clientes['Electrodoméstico'].fillna(Clientes['Valor Unitario'].map(valores_unitarios ))
Clientes.info()
```

```python
Clientes['Hora Compra'] = pd.to_datetime(Clientes['Hora Compra'], format='%H:%M:%S')
```

```python
Clientes.head(3)
```

```python
Clientes['Ano_de_compra'] = Clientes['Fecha Compra'].dt.year
Clientes['Mes_de_compra'] = Clientes['Fecha Compra'].dt.month
Clientes['Hora'] = Clientes['Hora Compra'].dt.hour
```

```python
Clientes.head(3)
```

```python
Categorias = {
    'Aspiradora': 'Aseo',
    'Cafetera': 'Cocina',
    'Horno eléctrico': 'Cocina',
    'Lavadora': 'Aseo',
    'Licuadora': 'Cocina',
    'Microondas': 'Cocina',
    'Nevera': 'Cocina',
    'Plancha':'Aseo',
    'Televisor': 'Confort',
    'Ventilador': 'Confort'
}
```

```python
Clientes['Categoria'] = Clientes['Electrodoméstico'].map(Categorias)
Clientes.head()
```

```python
Clientes['Valor_medio'] = Clientes['Valor Total']/Clientes['Cantidad']
Clientes.head()
```

```python
Clientes.to_excel('Clientes_transformados.xlsx', index=False)
```
