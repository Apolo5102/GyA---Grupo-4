```python
fecha_uso_actual = "2025-01"
```

```python
# Diccionario que mapea cada fuente con su DataFrame y la columna de invoice number
fuentes_map = {
    "duro last": (df_duro_last, "ORIGINAL Invoice", "PERIOD USED"),
    "fsbp": (df_fsbp, "Invoice", "Period Used"),
    "ses": (df_ses, "Invoice #", "Period Used"),
    "gaco": (df_gaco, "Invoice #", "Period Used"),
    "negatives": (df_negativos, "Document Number", "Return Period Used")
}
# Crear un DataFrame vac√≠o para almacenar las modificaciones
facturas_modificadas = pd.DataFrame()

# Iterar sobre cada fila de facturas_pagables
for index, row in facturas_pagables.iterrows():
    invoice_number = row["invoice_number"]
    fuente = row["fuente"]

    # Lista de posibles DataFrames a modificar
    if fuente == "fsbp":
        df_fsbp.loc[df_fsbp['Invoice'] == invoice_number, 'Period Used'] = fecha_uso_actual
        facturas_modificadas = pd.concat([facturas_modificadas, df_fsbp.loc[df_fsbp['Invoice'] == invoice_number]])
    elif fuente == "ses":
        df_ses.loc[df_ses['Invoice #'] == invoice_number, 'Period Used'] = fecha_uso_actual
        facturas_modificadas = pd.concat([facturas_modificadas, df_ses.loc[df_ses['Invoice #'] == invoice_number]])
    elif fuente == "gaco":
        df_gaco.loc[df_gaco['Invoice #'] == invoice_number, 'Period Used'] = fecha_uso_actual
        facturas_modificadas = pd.concat([facturas_modificadas, df_gaco.loc[df_gaco['Invoice #'] == invoice_number]])
    elif fuente == "negatives":
        df_negativos.loc[df_negativos['Document Number'] == invoice_number, 'Return Period Used'] = fecha_uso_actual
        facturas_modificadas = pd.concat([facturas_modificadas, df_negativos.loc[df_negativos['Document Number'] == invoice_number]])
    elif fuente == "duro last":
        df_duro_last.loc[df_duro_last['ORIGINAL Invoice'] == invoice_number, 'PERIOD USED'] = fecha_uso_actual
        facturas_modificadas = pd.concat([facturas_modificadas, df_duro_last.loc[df_duro_last['ORIGINAL Invoice'] == invoice_number]])
    else:
        print(f"Fuente no reconocida: {fuente}")

# Mostrar el DataFrame con todas las facturas modificadas
facturas_modificadas
```
